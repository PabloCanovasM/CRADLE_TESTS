```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambda_filename = "build/lambda_output.txt"

values = np.loadtxt(lambda_filename,dtype=np.float64)
values_eq = values[(np.indices(values.shape))[0] == (np.indices(values.shape))[1]]
values_less = values[((np.indices(values.shape))[0] - (np.indices(values.shape))[1]) == 2]
values_more = values[((np.indices(values.shape))[0] - (np.indices(values.shape))[1]) == -2]
print(values_eq,values_less,values_more)
j_max = (values.shape[0]-1)/2.
j = np.linspace(0,j_max,values.shape[0])
fine_j = np.linspace(0,j_max,8*(values.shape[0]-1)+1)


fig, ax = plt.subplots()

ax.set_xlabel("$J$")
ax.set_ylabel("$\\lambda_{J,J'}$")

ax.plot(j,values_eq,'ro',label="J'= J Code")
ax.plot(fine_j,1/(fine_j+1),'r-',label="J'= J Theory")

ax.plot(j[:-2],values_more,'bo',label="J'= J+1 Code")
ax.plot(fine_j,-fine_j/(fine_j+1),'b-',label="J'= J+1 Theory")

ax.plot(j[2:],values_less,'go',label="J'= J-1 Code")
ax.plot(j[2:],np.ones_like(j[2:]),'g-',label="J'= J-1 Theory")

fig.legend(bbox_to_anchor=(1.0,0.375,0.2,0.25))
fig.savefig("plots/lambda_test_result.png",bbox_inches="tight")

lambda_filename = "build/Lambda_output.txt"

values = np.loadtxt(lambda_filename,dtype=np.float64)
values_eq = values[(np.indices(values.shape))[0] == (np.indices(values.shape))[1]]
values_less = values[((np.indices(values.shape))[0] - (np.indices(values.shape))[1]) == 2]
values_more = values[((np.indices(values.shape))[0] - (np.indices(values.shape))[1]) == -2]
print(values_eq,values_less,values_more)
j_max = (values.shape[0]-1)/2.
j = np.linspace(0,j_max,values.shape[0])
fine_j = np.linspace(0,j_max,8*(values.shape[0]-1)+1)


fig, ax = plt.subplots()

ax.set_xlabel("$J$")
ax.set_ylabel("$\\Lambda_{J,J'}$")

ax.plot(j,values_eq,'ro',label="J'= J Code")
ax.plot(fine_j,-(2*fine_j-1)/(fine_j+1),'r-',label="J'= J Theory")

ax.plot(j[:-2],values_more,'bo',label="J'= J+1 Code")
ax.plot(fine_j,fine_j*(2*fine_j-1)/(fine_j+1)/(2*fine_j+3),'b-',label="J'= J+1 Theory")

ax.plot(j[2:],values_less,'go',label="J'= J-1 Code")
ax.plot(j[2:],np.ones_like(j[2:]),'g-',label="J'= J-1 Theory")

fig.legend(bbox_to_anchor=(1.0,0.375,0.2,0.25))
fig.savefig("plots/Lambda_test_result.png",bbox_inches="tight")
```

```python
def plot_params(data,At,Bt,Dt,x,points,oName):
    fig, ax = plt.subplots()

    ax.set_xlabel("$E$ (keV)")
    ax.set_ylabel("$Parameters$")

    ax.plot(data[::points,0],data[::points,1],'rx',label="A Code")
    ax.plot(x,At,'r-',label="A Theory")

    ax.plot(data[::points,0],data[::points,2],'bx',label="B Code")
    ax.plot(x,Bt,'b-',label="B Theory")

    ax.plot(data[::points,0],data[::points,3],'gx',label="D Code")
    ax.plot(x,Dt,'g-',label="D Theory")

    fig.legend(bbox_to_anchor=(1.0,0.375,0.2,0.25))
    fig.savefig("plots/"+oName,bbox_inches="tight")

data = np.loadtxt("build/ctcap_real_gt_12.txt")

alphaZ = 0.0072973525664*14
gamma = np.sqrt(1-alphaZ**2)
m_e = 510.9989461 #keV
lambda_jj = -0.5

x = np.arange(data[:,0].min(),data[:,0].max()+1,0.25)
B_theory = lambda_jj*gamma*m_e/x
A_theory = 0*x
D_theory = 0*x

plot_params(data,A_theory,B_theory,D_theory,x,40,"ctcap_real_gt_result.png")
```

```python
def plot_params_log(data,At,Bt,Dt,log_axis,x,points,oName):
    fig, ax = plt.subplots()

    ax.set_xlabel("$E$ (keV)")
    ax.set_ylabel("$Parameters$")
    ax.set_yscale("symlog",linthresh = 0.01,linscale=0.3, subs = [2, 3, 4, 5, 6, 7, 8, 9])
    ax.minorticks_on()
    
    if log_axis == 0:
        mask = np.geomspace(0.01, data.shape[0] - 1, num=50, endpoint=True, dtype="int_")
    else:
        mask = np.arange(data.shape[0])[::points]
    ax.plot(data[mask,0],data[mask,1],'rx',label="A Code")
    ax.plot(x,At,'r-',label="A Theory")

    if log_axis == 1:
        mask = np.geomspace(0.01, data.shape[0] - 1, num=50, endpoint=True, dtype="int_")
    else:
        mask = np.arange(data.shape[0])[::points]
    ax.plot(data[mask,0],data[mask,2],'bx',label="B Code")
    ax.plot(x,Bt,'b-',label="B Theory")

    if log_axis == 2:
        mask = np.geomspace(0.01, data.shape[0] - 1, num=50, endpoint=True, dtype="int_")
    else:
        mask = np.arange(data.shape[0])[::points]
    ax.plot(data[mask,0],data[mask,3],'gx',label="D Code")
    ax.plot(x,Dt,'g-',label="D Theory")

    fig.legend(bbox_to_anchor=(1.0,0.375,0.2,0.25))
    fig.savefig("plots/"+oName,bbox_inches="tight")

data = np.loadtxt("build/ctcap_comp_gt_12.txt")

lambda_jj = -0.5
x = np.arange(data[:,0].min(),data[:,0].max()+1,0.25)
v = np.sqrt(x**2/m_e**2-1)

A_theory = lambda_jj*alphaZ/v
B_theory = 0*x
D_theory = 0*x

plot_params_log(data,A_theory,B_theory,D_theory,0,x,40,"ctcap_comp_gt_result.png")
```

```python
def plot_params_log_single(data,a,log_axis,x,points,oName):
    fig, ax = plt.subplots()

    ax.set_xlabel("$E$ (keV)")
    ax.set_ylabel("$Parameters$")
    ax.set_yscale("symlog",linthresh = 0.01,linscale=0.3, subs = [2, 3, 4, 5, 6, 7, 8, 9])
    ax.minorticks_on()
    
    if log_axis == 0:
        mask = np.geomspace(0.01, data.shape[0] - 1, num=50, endpoint=True, dtype="int_")
    else:
        mask = np.arange(data.shape[0])[::points]
    ax.plot(data[mask,0],data[mask,1],'rx',label="a Code")
    ax.plot(x,a,'r-',label="a Theory")

    fig.legend(bbox_to_anchor=(1.0,0.375,0.2,0.25))
    fig.savefig("plots/"+oName,bbox_inches="tight")


data = np.loadtxt("build/cscv_comp_f.txt")

alphaZ = 0.0072973525664*14
gamma = np.sqrt(1-alphaZ**2)
m_e = 510.9989461 #keV
lambda_jj = -0.5

x = np.arange(data[:,0].min(),data[:,0].max()+1,0.25)
v = np.sqrt(x**2/m_e**2-1)

a_theory = -alphaZ/v

plot_params_log_single(data[:,:2],a_theory,0,x,40,"cscv_comp_fermi_result.png")
```

```python
def plot_params_log2(data,at,ct,log_axis,x,points,oName):
    fig, ax = plt.subplots()

    ax.set_xlabel("$E$ (keV)")
    ax.set_ylabel("$Parameters$")
    ax.set_yscale("symlog",linthresh = 0.01,linscale=0.3, subs = [2, 3, 4, 5, 6, 7, 8, 9])
    ax.minorticks_on()
    
    if log_axis == 0:
        mask = np.geomspace(0.01, data.shape[0] - 1, num=50, endpoint=True, dtype="int_")
    else:
        mask = np.arange(data.shape[0])[::points]
    ax.plot(data[mask,0],data[mask,1],'rx',label="a Code")
    ax.plot(x,at,'r-',label="a Theory")

    if log_axis == 1:
        mask = np.geomspace(0.01, data.shape[0] - 1, num=50, endpoint=True, dtype="int_")
    else:
        mask = np.arange(data.shape[0])[::points]
    ax.plot(data[mask,0],data[mask,2],'bx',label="c Code")
    ax.plot(x,ct,'b-',label="c Theory")

    fig.legend(bbox_to_anchor=(1.0,0.375,0.2,0.25))
    fig.savefig("plots/"+oName,bbox_inches="tight")
```
